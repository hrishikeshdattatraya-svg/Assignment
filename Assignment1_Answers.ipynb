## Question 1: What is Data Science?

### Definition
Data Science is an interdisciplinary field that combines mathematics and statistics, computer programming, and domain knowledge to extract meaningful insights and actionable decisions from data. It is the practice of collecting, cleaning, analyzing, and interpreting large and complex datasets to solve real-world problems. Data scientists use tools like Python, SQL, machine learning libraries, and data visualization to turn raw numbers into stories that drive decisions.

### Why It Is More Than Just Statistics or Programming
Statistics alone can describe what happened in historical data — but it does not automatically translate those findings into business decisions or build systems that act on predictions. Programming alone can automate tasks and process data efficiently, but without statistical understanding or domain knowledge, the code produces numbers with no meaning.

Data Science sits at the intersection of all three: statistical thinking to model uncertainty, programming to handle scale and automation, and domain expertise to ask the right questions and interpret results in context.

### Real-World Example: Netflix Recommendation System
Netflix collects data on what millions of users watch, pause, rewind, rate, and abandon. Data Science combines statistics (collaborative filtering algorithms), programming (scalable Python and Spark pipelines), and domain knowledge about viewer behavior to build a recommendation engine. This system is estimated to save Netflix over $1 billion annually in customer retention.

## Question 5: Python Data Types & String Operations

### Part 1: Data Types of Each Variable
- `price = 499` → **int** (integer). Python assigns integer type to whole numbers with no decimal point.
- `discount = 0.15` → **float**. Any number with a decimal point is a float.
- `final_price = price - price * discount` → **float**. When an int is multiplied by a float (499 × 0.15 = 74.85), the result is always a float. So `final_price = 424.15`.

````python
# Question 5 – Python Data Types & String Operations

price = 499
discount = 0.15
final_price = price - price * discount

# Part 1: Check types
print(f'Type of price      : {type(price)}')
print(f'Type of discount   : {type(discount)}')
print(f'Type of final_price: {type(final_price)}')

# Part 2: Print final_price with 2 decimal places
print(f'Final Price: {final_price:.2f}')
# Output: Final Price: 424.15

# Part 3: Convert price to string and concatenate
# Method 1: Using str() and concatenation
message = 'The final price is Rs.' + str(round(final_price, 2))
print(message)

# Method 2: Using f-string (cleaner)
message = f'The final price is Rs.{final_price:.2f}'
print(message)
# Output: The final price is Rs.424.15
````

## Question 6: Bonus Calculator with Conditional Logic

The code uses:
- **Comparison operators** (`>=`): checks if the score meets a threshold.
- **`if/elif/else`**: creates a decision tree — only one branch executes.
- Checking `>= 90` first ensures a score of 95 doesn't incorrectly receive only 10% bonus.

**Sample Outputs:**
- Score = 95 → 20% bonus → Rs. 10000.00
- Score = 80 → 10% bonus → Rs. 5000.00
- Score = 60 → No bonus  → Rs. 0.00

````python
# Question 6 – Bonus Calculator with Conditional Logic

salary = 50000  # Base salary in rupees
score = int(input('Enter performance score: '))

if score >= 90:
    bonus_percent = 0.20
    bonus = salary * bonus_percent
    print(f'Score: {score} >= 90 -> 20% bonus')
    print(f'Bonus Amount: Rs.{bonus:.2f}')

elif score >= 75:
    bonus_percent = 0.10
    bonus = salary * bonus_percent
    print(f'Score: {score} >= 75 -> 10% bonus')
    print(f'Bonus Amount: Rs.{bonus:.2f}')

else:
    bonus = 0
    print(f'Score: {score} < 75 -> No bonus')
    print('Bonus Amount: Rs.0.00')
````

## Question 7: String Operations on Customer Review

**Given:** `review = "The product quality is Excellent but delivery was Late"`

**Key Notes:**
- `.lower()` is applied before checking for `'excellent'` to ensure case-insensitive search.
- `.split()` splits on whitespace; `len()` counts the resulting words.
- The `in` keyword checks substring membership and returns `True` or `False`.

````python
# Question 7 – String Operations on Customer Review

review = 'The product quality is Excellent but delivery was Late'

# 1. Convert to lowercase
review_lower = review.lower()
print('Lowercase:', review_lower)

# 2. Count number of words
word_count = len(review_lower.split())
print('Word Count:', word_count)

# 3. Check if 'excellent' exists
has_excellent = 'excellent' in review_lower
print('Contains excellent:', has_excellent)

# 4. Replace 'Late' with 'delayed'
updated_review = review.replace('Late', 'delayed')
print('Updated Review:', updated_review)
````

## Question 8: Sales Data Analysis with Lists

**Given:** `sales = [12000, 15000, 11000, 18000, 17000]`

Operations performed:
1. Total revenue using `sum()`.
2. Average revenue using `total / len()`.
3. Highest sales month using `max()` and `.index()` (+1 since months start from 1).
4. Append new month's sale using `.append()`.
5. Sort descending using `.sort(reverse=True)`.

````python
# Question 8 – Sales Data Analysis with Lists

sales = [12000, 15000, 11000, 18000, 17000]

# 1. Total revenue
total_revenue = sum(sales)
print('Total Revenue: Rs.', total_revenue)

# 2. Average revenue
average_revenue = total_revenue / len(sales)
print('Average Revenue: Rs.', average_revenue)

# 3. Highest sales month
max_sale = max(sales)
max_month = sales.index(max_sale) + 1
print(f'Highest Sales: Rs.{max_sale} in Month {max_month}')

# 4. Append new month
sales.append(20000)
print('Updated Sales List:', sales)

# 5. Sort descending
sales.sort(reverse=True)
print('Sorted (Descending):', sales)
````

## Question 9: Sales Analysis System

**Given:**
```python
products   = ["Laptop", "Phone", "Tablet"]
prices     = [50000, 20000, 15000]
quantities = [5, 10, 7]
```

The solution computes revenue per product (`price × quantity`), identifies the top product, and prints a formatted summary report.

````python
# Question 9 – Sales Analysis System

products   = ['Laptop', 'Phone', 'Tablet']
prices     = [50000, 20000, 15000]
quantities = [5, 10, 7]

# 1. Compute total revenue per product
revenues = []
for i in range(len(products)):
    revenue = prices[i] * quantities[i]
    revenues.append(revenue)

# 2. Identify product with maximum revenue
max_revenue = max(revenues)
max_index   = revenues.index(max_revenue)
top_product = products[max_index]

# 3. Print formatted summary
print('=' * 40)
print('       SALES SUMMARY REPORT')
print('=' * 40)
for i in range(len(products)):
    print(f'{products[i]:<10} | Price: Rs.{prices[i]:>6} | Qty: {quantities[i]} | Revenue: Rs.{revenues[i]:>8}')
print('-' * 40)
print(f'Total Revenue: Rs.{sum(revenues)}')
print(f'Top Product: {top_product} with Rs.{max_revenue}')
print('=' * 40)
````

## Question 10: Customer Feedback System — End-to-End Explanation

### Step 1: How Raw Text Becomes Structured Data
A customer review like *'Delivery was very slow and packaging was damaged'* is raw unstructured text. NLP preprocessing steps transform it:
- **Text Cleaning**: Remove punctuation, special characters, and extra spaces.
- **Lowercasing**: Convert all text to lowercase.
- **Tokenization**: Split into individual words (tokens).
- **Stop Word Removal**: Remove words like 'was', 'and', 'the' that carry no sentiment signal.
- **Feature Extraction**: Convert tokens into numerical vectors using TF-IDF or embeddings.
- **Label Assignment**: Pair each review vector with the customer's star rating.

### Step 2: Where Python (Strings & Lists) is Used
- String operations (`.lower()`, `.replace()`, `.split()`, `.strip()`) clean and normalize text.
- Lists store tokens, hold stop words, and contain all cleaned reviews and ratings.
- List comprehensions efficiently apply cleaning across hundreds of thousands of reviews.
- F-strings display processed results to analysts.

### Step 3: Where Data Science Models are Used
- **Sentiment Classification**: Logistic Regression, Random Forest, or BERT classifies reviews as Positive, Neutral, or Negative.
- **Topic Modeling**: LDA (Latent Dirichlet Allocation) identifies recurring themes such as 'delivery issues' or 'product quality'.
- **Rating Prediction**: A regression model predicts a numerical rating from review text.
- **Anomaly Detection**: Statistical models flag unusual spikes in negative reviews.

### Step 4: Where GenAI Can Enhance the System
- **Automated Response Generation**: An LLM generates personalized, empathetic replies to each customer review.
- **Insight Summarization**: A GenAI model summarizes top complaints and compliments into a weekly report.
- **Review Categorization with Context**: LLMs understand nuanced language (sarcasm, regional expressions) better than keyword-based classifiers.
- **Chatbot for Follow-up**: A GenAI-powered chatbot proactively resolves issues for customers who left negative reviews.

### Step 5: Ethical Concerns That Must Be Addressed
- **Privacy and Consent**: Customers must consent to AI analysis; data must be anonymized.
- **Bias in Sentiment Models**: Training data dominated by one demographic may produce biased insights.
- **Transparency**: Customers must know when they are interacting with AI.
- **Data Security**: Feedback data must be stored and transmitted securely.
- **Misuse of Insights**: Sentiment data must not be used for manipulative targeted pricing or psychological persuasion.
